```python
x = symbols('x', real = True)
n = symbols('n', integer = True)

an = ((-2+(4*x**2))**n)/((4**n)*n)
ap = ((-2+(4*x**2))**(n+1))/((4**(n+1))*(n+1))
term = ap/an
simp = simplify(term)

lim = limit(simp, n, oo)
rc = solve(lim-1, x)

first = simplify(an.subs(x, -3/4))
second = simplify(an.subs(x, 5/4))


print("Radius of Convergence =", rc[1])
print()
print("Endpoints are -3/4 and 5/4")
print()
print("Substituting -3/4 into the series gives", first, "which converges by geometric series with r being 1/16 which is < 1")
print() 
print("Substituting 5/4 into the series gives", second, 'which diverges by geometric series since', second,"> 1")
print()
print("Therefore, the interval of convergence is", "[-3/4, 5/4)")
```

### Question 3

#### 3a

```python
# Radius of convergence of the series
# Ratio test


n = symbols("n",integer=True)
x = symbols("x",real=True)


an = (((-1) ** n) *(x ** ((2 * n) + 1))) / (factorial(n) * factorial (n + 1) * (2 ** ((2 * n) + 1)))


RatioTest = abs(an.subs(n,n+1)/an)


print('The Ratio Test is',RatioTest,'which simplifies to',RatioTest.simplify())


L = limit(RatioTest,n,oo)


print('The limit of the Ratio Test is',L,'which is always < 1 so the radius of convergence is oo and the interval is (-oo,oo)')
```
